```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import warnings
from tensorflow.examples.tutorials.mnist import input_data


warnings.filterwarnings('ignore')
plt.style.use('ggplot')
%matplotlib inline
```

### 定义一个命名空间

```python
location = 'E:/Dataset/MNIST/'
mnist = input_data.read_data_sets(train_dir=location,one_hot=True)

def variable_summaries(var):
    with tf.name_scope('summaries'):
        mean = tf.reduce_mean(var)
        tf.summary.scalar('mean',mean)
        with tf.name_scope('stddev'):
            stddev = tf.sqrt(tf.reduce_mean(tf.square(var-mean)))
        tf.summary.scalar('stddev',stddev)
        tf.summary.scalar('max',tf.reduce_max(var))#最大值
        tf.summary.scalar('min',tf.reduce_min(var))
        tf.summary.histogram('histogram',var)

        
#命名空间
with tf.name_scope('input'):
    x = tf.placeholder(tf.float32,[None,784],name='x-input')
    y_label = tf.placeholder(tf.float32,[None,10],name='y_label-input')
    
with tf.name_scope('layer'):
    with tf.name_scope('weights'):
        w = tf.Variable(tf.zeros([784,10]),name='first_weight')
        variable_summaries(w)
    with tf.name_scope('biases'):
        b = tf.Variable(tf.zeros([10]),name='Bias')
        variable_summaries(b)
    with tf.name_scope('linear_y'):
        y = tf.matmul(x,w) + b
#线性输出



#----------超参数-------------------------------
learning_rate = 0.1
batch_num = 50
batch_size = 100
#----------超参数-------------------------------


with tf.name_scope('loss1'):
    cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=y,labels=y_label))
    tf.summary.scalar('loss',cross_entropy)
with tf.name_scope('train'):
    #采用SGD优化器
    train_step = tf.train.GradientDescentOptimizer(learning_rate).minimize(cross_entropy)

    
#-----------定义准确率计算方法----------------评估模型
with tf.name_scope('accuracy'):
    with tf.name_scope('correct_prediction'):
        correction_prediction = tf.equal(tf.argmax(y,1),tf.argmax(y_label,1))
    with tf.name_scope('accuracy'):
        accuracy = tf.reduce_mean(tf.cast(correction_prediction,tf.float32))  #将布尔型转化为浮点型
        tf.summary.scalar('accuracy',accuracy)

        
#合并所有的summary
merged = tf.summary.merge_all()
    
init = tf.global_variables_initializer()
#--------------------------------------------------------------------
#                开启会话，计算图
#--------------------------------------------------------------------
with tf.Session() as sess:
    
    sess.run(init)
    writer = tf.summary.FileWriter('logs/',sess.graph)
    
    for epoch in range(200):
        for i in range(batch_num):
            batch  =  mnist.train.next_batch(batch_size)
            summary,_ = sess.run([merged,train_step],feed_dict={x:batch[0], y_label:batch[1]})
            
        writer.add_summary(summary,epoch)
        #----------在测试集上计算准确率----------------------
        test_accuracy = sess.run(accuracy,feed_dict={x:mnist.test.images, y_label: mnist.test.labels})
        print('模型在测试集上的准确率为：',test_accuracy)
```
